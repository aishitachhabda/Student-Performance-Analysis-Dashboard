```python
%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

st.set_page_config(layout="wide")

# ---------------- DATA ----------------
np.random.seed(42)

students = ["Aarav","Priya","Rohit","Sneha","Karan",
            "Divya","Arjun","Ananya","Vivek","Pooja"]

subjects = ["Math","Physics","Chemistry","Programming","English"]

data = []
for s in students:
    for sub in subjects:
        e1,e2,e3,e4 = np.random.randint(40,100,4)
        att = np.random.randint(60,100)

        avg = (e1+e2+e3+e4)/4
        grade = "A" if avg>=85 else "B" if avg>=70 else "C" if avg>=55 else "D"

        data.append([s,sub,e1,e2,e3,e4,att,avg,grade])

df = pd.DataFrame(data, columns=[
    "Student","Subject","Exam1","Exam2","Exam3","Exam4",
    "Attendance","Average","Grade"
])


st.sidebar.title("🎛️ Filters")

student = st.sidebar.selectbox("Select Student", ["All"] + students)
subject = st.sidebar.multiselect("Select Subjects", subjects, default=subjects)
grade = st.sidebar.multiselect("Select Grades", ["A","B","C","D"], default=["A","B","C","D"])

filtered_df = df.copy()

if student != "All":
    filtered_df = filtered_df[filtered_df["Student"] == student]

filtered_df = filtered_df[
    (filtered_df["Subject"].isin(subject)) &
    (filtered_df["Grade"].isin(grade))
]

st.title("🎓 Student Performance Dashboard")


col1,col2,col3 = st.columns(3)

col1.metric("Average Marks", round(filtered_df["Average"].mean(),1))
col2.metric("Top Score", round(filtered_df["Average"].max(),1))
col3.metric("Attendance", round(filtered_df["Attendance"].mean(),1))

st.subheader("📊 Subject Performance")

fig1 = px.bar(filtered_df, x="Subject", y="Average", color="Average")
st.plotly_chart(fig1, use_container_width=True)

st.subheader("📈 Exam Trend")

exam_df = filtered_df.melt(
    id_vars=["Subject"],
    value_vars=["Exam1","Exam2","Exam3","Exam4"],
    var_name="Exam", value_name="Marks"
)

fig2 = px.line(exam_df, x="Exam", y="Marks", color="Subject", markers=True)
st.plotly_chart(fig2, use_container_width=True)

st.subheader("🎯 Grade Distribution")

fig3 = px.pie(filtered_df, names="Grade")
st.plotly_chart(fig3)

st.subheader("🔥 Student Performance Heatmap")

pivot = df.pivot_table(
    index="Student",
    columns="Subject",
    values="Average"
)

pivot = pivot.fillna(0)

fig, ax = plt.subplots(figsize=(8,5))

sns.heatmap(
    pivot,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    fmt=".1f"
)

plt.xticks(rotation=30)
plt.yticks(rotation=0)

st.pyplot(fig)

st.subheader(" Prediction (Linear Regression)")

X = filtered_df[["Exam1","Exam2","Exam3","Attendance"]]
y = filtered_df["Exam4"]

if len(filtered_df) > 5:
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

    model = LinearRegression()
    model.fit(X_train,y_train)

    pred = model.predict(X_test)

    pred_df = pd.DataFrame({
        "Actual": y_test.values,
        "Predicted": pred
    })

    st.dataframe(pred_df.head())
else:
    st.warning("Not enough data for prediction")

# ---------------- INSIGHTS ----------------
st.subheader("🚨 Insights")

avg_student = df.groupby("Student")["Average"].mean()

top = avg_student.idxmax()
low = avg_student.idxmin()

st.success(f" Top Performer: {top}")
st.error(f" Needs Improvement: {low}")

st.write(" Suggestion: Focus on weak subjects and maintain attendance.")
```
